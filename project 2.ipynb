## Intermediate level

#### Count how many movies have each of the content ratings.

#### Use a visualization to display that same data, including a title and x and y labels.

```python
# Answer:
ratings = movies['content_rating'].value_counts().plot(kind='bar');

#set the title
ratings.set_title('Number of Films by Rating');

#set the x-axis
ratings.set_xlabel('Content Ratings');

#set the y-axis
ratings.set_ylabel('Number of Films');
```

#### Convert the following content ratings to "UNRATED": NOT RATED, APPROVED, PASSED, GP.

```python
# Answer:
movies = movies.replace(['NOT RATED','APPROVED','PASSED','GP'],'UNRATED')
movies[movies['content_rating']=='UNRATED']
```

#### Convert the following content ratings to "NC-17": X, TV-MA.

```python
# Answer:
movies = movies.replace(['X','TV-MA'],'NC-17')
movies[movies['content_rating']=='NC-17']
```

#### Count the number of missing values in each column.

```python
# Answer:
movies.isnull().sum(axis = 0)
```

#### If there are missing values: examine them, then fill them in with "reasonable" values.

```python
# Answer:
movies[movies.isnull().any(axis=1)] #find all the rows with NaN
```

```python
#Since it's only three missing values I can check their actual content ratings by hand
movies.loc[movies['title'] == 'Butch Cassidy and the Sundance Kid', 'content_rating'] = 'PG'
movies.loc[movies['title'] == 'Where Eagles Dare', 'content_rating'] = 'PG'

#There are two movies with the title "True Grit". I had to filter on the title and duration to get the one I wanted.
movies.loc[(movies['title'] == 'True Grit') & (movies['duration'] == 128), 'content_rating' ] = 'G'
```

```python
movies[movies['title']=='Butch Cassidy and the Sundance Kid']
movies[movies['title']=='Where Eagles Dare']
movies[movies['title']=='True Grit']
```

#### Calculate the average star rating for movies 2 hours or longer, and compare that with the average star rating for movies shorter than 2 hours.

```python
# Answer: Movies longer than 2 hours have a slightly higher star rating
movies[movies['duration']>120]['star_rating'].mean()
movies[movies['duration']<120]['star_rating'].mean()
```

#### Use a visualization to detect whether there is a relationship between duration and star rating.

```python
# Answer:
sns.pairplot(movies);
movies.plot(kind='scatter', x='duration', y='star_rating', alpha=0.2)
```

#### Calculate the average duration for each genre.

```python
# Answer:
movies.groupby('genre')['duration'].mean()
```
